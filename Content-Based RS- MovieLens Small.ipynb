## Asking User Ratings

```python
def is_in_file(fav__id, fav__rating):
    if(fav__id<1):
        print("Movie ID needs to start with 1!")
        return False

    if(fav__id not in df['movieId'].values):
        print("This movie ID does not exist!")
        return False

    if(fav__id in fav_id):
        print("You have already rated this!")
        return False

    if(fav__rating>10):
        print("Ratings must be below 10!")
        return False
    else:
        fav_id.append(fav__id)
        fav_rating.append(fav__rating)
        return True
```

```python
fav_id=[]
fav_rating=[]

while(1):
    fav__id=int(input('Please enter movie #1 id: '))
    fav__rating=int(input('Please enter movie #1 rating out of 10: '))

    if(is_in_file(fav__id,fav__rating))==True:
      break

print("\n")

while(1):
    fav__id=int(input('Please enter movie #2 id: '))
    fav__rating=int(input('Please enter movie #2 rating out of 10: '))

    if(is_in_file(fav__id,fav__rating))==True:
      break

print("\n")

while(1):
    fav__id=int(input('Please enter movie #3 id: '))
    fav__rating=int(input('Please enter movie #3 rating out of 10: '))

    if(is_in_file(fav__id,fav__rating))==True:
      break


print("\n\n~~~~~~~~RATED MOVIES~~~~~~~~~")

for i in range(0, 3):
    name = df[df['movieId'] == fav_id[i]]['title'].iloc[0]
    #rating = fav_rating[i]
    print("Movie ID: {:<8} Movie Name: {:<50} Rating: {:<1}/10".format(fav_id[i], name, fav_rating[i]))


```

```python
movie_genres=[]

for i in range(0,3):
  genres=df[df['movieId'] == fav_id[i]]['genres'].iloc[0].split('|')

  for genre in genres:
    if genre not in movie_genres:
      movie_genres.append(genre)


print(movie_genres)
```

```python
#user profile is a dictionary with key as the genre and value as the normalized movie matrix
user_profile={genre:0 for genre in movie_genres}
#print(user_profile)

genres_mov1={genre:1 if genre in df[df['movieId']==fav_id[0]]['genres'].iloc[0].split('|') else 0 for genre in movie_genres}
genres_mov2={genre:1 if genre in df[df['movieId']==fav_id[1]]['genres'].iloc[0].split('|') else 0 for genre in movie_genres}
genres_mov3={genre:1 if genre in df[df['movieId']==fav_id[2]]['genres'].iloc[0].split('|') else 0 for genre in movie_genres}

print("\n~~~~~~ MOVIE MATRIX ~~~~~~")
print(genres_mov1)
print(genres_mov2)
print(genres_mov3)


#now we must multiply each value in dict with the rating we gave initially
print("\n\n")

genres_mov1={key: value*fav_rating[0] for key,value in genres_mov1.items()}
genres_mov2={key: value*fav_rating[1] for key,value in genres_mov2.items()}
genres_mov3={key: value*fav_rating[2] for key,value in genres_mov3.items()}

print("\n~~~~~~ MOVIE MATRIX x USER RATINGS ~~~~~~")
print(genres_mov1)
print(genres_mov2)
print(genres_mov3)


```

## Calculating User Profile

```python
#now we need to normalize the movie matrix to get the user profile
norm_sum=[]
for key in genres_mov1.keys():
  norm_sum.append(genres_mov1[key] + genres_mov2[key] + genres_mov3[key])


#print(norm_sum)
#print(sum(norm_sum))
total = sum(norm_sum)

for i in range(0,len(norm_sum)):
  norm_sum[i]=round(norm_sum[i]/total,2)

print(norm_sum)


keys = list(user_profile.keys())

for i in range(len(keys)):
        user_profile[keys[i]] = norm_sum[i]


print("\n~~~~~~ USER PROFILE ~~~~~~")
print(user_profile)
```

## Creating a matrix/dict with every movie and its rating in those specific genres

```python
len(df['movieId'])
```

```python
#were going to pick top 5 from results
results=[]
movieID_results=[] #going to store corresponding movieIDs here, so we can select which 5 movies was rated higest later on

#traversing each movie in csv file
for movieID in df['movieId']:

  #except for the movies we have already chosen
  if movieID not in fav_id:
    #separating their genres
    genres = df[df['movieId'] == movieID]['genres'].iloc[0].split('|')

    #populating a genre vector compared to genres of the movies we have already picked
    genre_vector = [1 if genre in genres else 0 for genre in movie_genres]

    #multiplying it with our user profile (i eneded up using normalized list rather than dict)
    res_list = [genre_vector[i] * norm_sum[i] for i in range(len(norm_sum))]

    #putting the sum of each movie into results matrix
    results.append(round(sum(res_list),2))
    movieID_results.append(movieID)


print(results)
print(movieID_results)
```

```python
#now we convert these ratings out of 10

results=[results[i]*10 for i in range(0,len(results))]
print(results)
```

```python
for i in range(0,len(results)):
  results[i]=round(results[i],2)
```

```python
#now im going to create a new dictionary with keys as movieID results
#and the values of that dict as the calculated results/ratings of other movies
#that we did not select in our 3 (unrated movies)

new_dict = {key: value for key, value in zip(movieID_results, results)}
print(new_dict)
```

## These are the 5 recommended movies based on the user profile genre matrix
You can correlate the genres of these 5 movies (from CSV) with the user profile matrix genres.


```python
#now we just need to sort the dictionary in reverse order to find top 5 recommendable movies
#i wont lie, i copied this portion from chatGPT because i hate working with dictionaries :(

sorted_ratings = sorted(new_dict.items(), key=lambda x: x[1], reverse=True)

# Displaying the top 5 movie IDs with the highest ratings
top_5 = sorted_ratings[:5]
for movieID, rating in top_5:
    name = df[df['movieId'] == movieID]['title'].iloc[0]
    print("Movie ID: {:<8} Movie Name: {:<50} Rating: {:<1}/10".format(movieID, name, rating))

```
